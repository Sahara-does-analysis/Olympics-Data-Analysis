```python
df.dropna(inplace=True)
```

```python
df['Age'] = df['Age'].astype('int')
```

```python
df.info()
```

```python
df.columns
```

```python
df.rename(columns = {'Unnamed: 0':'Number',
                     'Sex':'Gender'},inplace = True)
```

```python
df.describe()
```

```python
df.columns
```

```python
df["Gender"] = df["Gender"].replace({'Male':'M','Female':'F'})
```

```python
 df.groupby(['Gender'],as_index= False)['Age'].count().sort_values(by='Age', ascending= False)
```

```python
Gender_ana= sns.countplot(x= 'Gender', data= df)

for bars in Gender_ana.containers:
    Gender_ana.bar_label(bars)
```

# CONCLUSION :-

## FROM THE ABOVE GRAPH WE CAN CONCLUDE THAT THE PARTICIPATION OF MALE ATHLETES IN HISTORY OF OLYMPICS IS ALWAYS GREATER THAN THE PARTICIPATION OF FEMALE ATHLETES. 

```python
df['Medal'] = df['Medal'].replace({'0' : 'None',
                                   '1' : 'Bronze',
                                   '2' : 'Silver',
                                   '3' : 'Gold'})
```

```python
medal_tally= sns.countplot(x= 'Medal', data= df)

for bars in medal_tally.containers:
    medal_tally.bar_label(bars)
```

# CONCLUSION :-

## FROM ABOVE GRAPH IT IS CLEAR THAT THIRTEEN THOUSAND EIGHT HUNDERED AND NINE ATHLETES EARNED GOLD MEDAL, THIRTEEN THOUSAND THREE HUNDERED AND NINTEEN EARNED SILVER MEDAL AND THIRTEEN THOUSAND FIVE HUNDERED AND FOURTY FOUR ATHLETES EARNED BRONZE MEDAL. 

```python
sns.countplot(data= df, x= "Age", hue= 'Gender')

sns.set(rc={'figure.figsize':(20,10)})
```

# CONCLUSION :-

## FROM ABOVE GRAPH WE CAN OBSERVE THAT THE ATHLETES OF AGE GROUP 18-30 YEARS HAVE MAJORLY PARTICIPATED IN THE OLYMPICS. WE CAN ALSO SEE THAT THE GRAPH IS POSITIVELY SKEWED.

```python
top_10_sports = df.Sport.value_counts().sort_values(ascending = False).head(10)
top_10_sports
```

```python
plt.figure(figsize = (50,10))
plt.subplot()
plt.pie(top_10_sports, labels= ['Athletics', 'Gymnastics', 'Swimming', 'Shooting', 'Cycling', 'Rowing', 'Fencing', 'Cross Country Skiing', 'Alpine Skiing', 'Wrestling'], autopct= '%1.1f%%')
plt.title('Top 10 Sports Distribustion')
plt.show()
```

# CONCLUSION :-

## IN THE ABOVE GRAPH THE DISTRIBUSTION OF ONLY 10 MOSTLY PLAYED SPORTS IS SHOWN, FROM THIS INFORMATION WE SAY THAT GLOBALLY ATMOST PLAYED SPORTS ARE ATHLETICS, GYMNASTICS AND SWIMMING... 

```python
sns.set(rc={'figure.figsize':(15,10)})
Season_ana= sns.countplot(x= 'Season', data= df)

for bars in Season_ana.containers:
    Season_ana.bar_label(bars)
```
